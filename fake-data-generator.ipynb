# 목표: 실습 데이터를 위한 가짜 데이터 만들기

```python
# 가짜 데이터를 만들기 위해서 fake-factory 라는 라이브러리를 사용할 예정입니다.

from faker import Faker


fake = Faker('ko_KR')

print(fake.name())
print(fake.address())
```

```python
import random

import pandas as pd
```

```python
zigbang_df = pd.DataFrame(columns=[
    "address",
    "phonenumber",
    "deposit",
    "rent",
])


for _ in range(1000):
    zigbang_df.loc[len(zigbang_df)] = [
        fake.address(),
        "010" + "".join([str(random.choice(range(1, 10))) for x in range(8)]),
        random.choice(range(1, 10)) * 1000,
        random.choice(range(3, 8)) * 10,
    ]
```

```python
zigbang_df
```

```python
dabang_df = pd.DataFrame(columns=[
    "address",
    "phonenumber",
    "deposit",
    "rent",
])


for _ in range(700):
    dabang_df.loc[len(dabang_df)] = [
        fake.address(),
        "010" + "".join([str(random.choice(range(1, 10))) for x in range(8)]),
        random.choice(range(1, 10)) * 1000,
        random.choice(range(3, 8)) * 10,
    ]
```

```python
dabang_df
```

```python
zigbang_df.to_csv("zigbang.csv", index=False)
dabang_df.to_csv("dabang.csv", index=False)
```
